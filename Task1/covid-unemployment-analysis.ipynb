# COVID-19 Unemployment Analysis

Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force.
We have seen a sharp increase in the unemployment rate during COVID-19, therefore this notebook focusses on analyzing unemployment rate. 

We're going to take the following approach:

1. Problem Definition
2. Data
3. Evaluation
4. Features
5. Data Visualisation
6. Experimentation

## 1. Problem Definition

> How well can we predict the unemployment analysis, given it's characterisitics during Covid-19 in India.

## 2. Data

The original data came from - https://unemploymentinindia.cmie.com/  <br>

This dataset was downloaded from **Kaggle - Unemployment in India.** <br>
https://www.kaggle.com/datasets/gokulrajkmv/unemployment-in-india

## 3. Evaluation

To figure out the top 5 states which were most affected by the unemployment.

## 4. Features

#### Data Dictionary:

- **Region**: States in India.
- **Date**: Date on which the unemployment rate was observed.
- **Frequency**: Measuring frequency (Monthly).
- **Estimated Unemployment Rate (%)**: Percentage of unemployed people in each state of India.
- **Estimated Employed (%)**: Percentage of employed people.
- **Estimated Labour Participation Rate (%)**: Labour force participation rate, calculated by dividing the number of people actively participating in the labor force by the total number of people eligible to participate in th labor.

In both the datasets, we convert the 'Date' column to datetime and 'Frequency' and 'Area' columns to categorical columns

```python
# Before converting into categories
data1.info()
```

```python
# Coverting to categories in data1
data1['Date']=pd.to_datetime(data1['Date'],dayfirst=True)
data1['Frequency'] = data1['Frequency'].astype('category')
data1['Area'] =data1['Area'].astype('category')
data1.info()
```

```python
# Before converting into categories
data2.info()
```

```python
# Converting data2 into categories
data2['Date']= pd.to_datetime(data2['Date'], dayfirst=True)
data2['Frequency'] = data2['Frequency'].astype('category')
data2['Region'] = data2['Region'].astype('category')
data2.info()
```

##### We extract the Month from the Date in a new column 'Month' in both the datasets

```python
# Create new month column 
data1['Month']=data1['Date'].dt.month

# Convert Month to Month Name
data1['Month Name']=data1['Month'].apply(lambda x: calendar.month_abbr[x])
data1
```

```python
# Create new month column 
data2['Month']=data2['Date'].dt.month

# Convert Month to Month Name
data2['Month Name']=data2['Month'].apply(lambda x: calendar.month_abbr[x])
data2
```

##### We will split the dataset data1 into two parts: <br> 
* One before lockdown Date Range: 2019-05-31' to '2020-03-31'
* and second one 2020-04-01 onwards.

```python
# We create a boolean variable 'mask' which will return true if it matches the date range 
mask1 = (data1['Date'] >= '2019-05-31') & (data1['Date'] <= '2020-03-31')
data1_new = data1.loc[mask1]
data1_new
```

## 5. Data Visualisation

```python
plt.figure(figsize=(30,10))
plt.subplot(1,2,1)
plt.title("Correlation BEFORE Lockdown")
sns.heatmap(data=data1_stats.corr(),annot=True,cmap='coolwarm')

plt.subplot(1,2,2)
plt.title("Correlation During Lockdown")
sns.heatmap(data=data2_stats.corr(),annot=True,cmap='coolwarm')
plt.show()
```

Plotting Graphs to Visualize distribution between Rural and Urban areas (Dataset 1) and Different regions(dataset 2)


```python
plt.figure(figsize=(25,10))

#Graph 1
plt.subplot(1,2,1)

sns.barplot(data=data1_new,x='Area',y='Estimated Unemployment Rate (%)',hue='Area',ci=None)

plt.xlabel("Area",fontsize=16)
plt.ylabel("Estimated Unemployment Rate (%)", fontsize=16)
plt.title("Unemployment before lockdown(Area wise)",fontsize=16)

#Graph 2
plt.subplot(1,2,2)

sns.barplot(data=data2_new, x='Region', y='Estimated Unemployment Rate (%)',ci=None,palette='Paired_r')

plt.xticks(rotation=90)
plt.xlabel("Region",fontsize=16)
plt.ylabel("Estimated Unemployment Rate (%)", fontsize=16)
plt.title("Unemployment after lockdown(Region wise)",fontsize=16)
```

Plotting State-wise Bar graph to find Unemployment Rate before and During lockdown

```python
#Statewise Stats
state_stats1=round(data1_new.groupby(['State'])
                   [['Estimated Unemployment Rate (%)','Estimated Employed',
                                         'Estimated Labour Participation Rate (%)']].mean().reset_index(),2)
state_stats1=state_stats1.sort_values(by='Estimated Unemployment Rate (%)',ascending=False)
```

```python
state_stats2=round(data2_new.groupby(['State'])
                   [['Estimated Unemployment Rate (%)','Estimated Employed',
                                         ' Estimated Labour Participation Rate (%)']].mean().reset_index(),2)
state_stats2=state_stats2.sort_values(by='Estimated Unemployment Rate (%)',ascending=False)
```

```python
state_stats1.head()
```

```python
state_stats2.head()
```

```python
plt.figure(figsize=(16,9))

#Before Lockdown
plt.subplot(1,2,1)
plt.title("State wise Unemployment Rate before lockdown")
plt.xlabel("States")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.xticks(rotation=90)

sns.barplot(data=state_stats1,x='State',y='Estimated Unemployment Rate (%)',ci=None)


#During lockdown
plt.subplot(1,2,2)
plt.title("State wise Unemployment Rate During lockdown")
plt.xlabel("States")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.xticks(rotation=90)

sns.barplot(data=state_stats2,x='State',y='Estimated Unemployment Rate (%)',ci=None)
```
